```python
import pandas as pd

pd.set_option('display.max_row', 100)

def frac_dilution(frac, max_conc, n=8):
    l = [max_conc]
    for i in range(n):
        l.append(l[-1] * frac)
    return l

def v1(v2, c1, c2):
    return (c2 * v2) / c1

cpds = pd.read_csv('selection.csv').drop(['Unnamed: 0'],
                                        axis=1)

cpds
```

```python
mws = {'tembotrione': 440.8,
      'fomesafen': 438.8,
      'atrazine': 215.68,
      'imazethapyr':289.33,
      'butafenacil':474.8,
      'nicosulfuron': 410.4,
      'dicamba': 221.03,
      'mesotrione': 339.32,
      'fluazifop': 327.25} # g / mol

def mgml(mw, c2): # c2: M
    # mw * c2 = g 
    return mw * c2 # mg / ml


for i in mws:
    print(i, mgml(mws[i], 0.01), ' mg/ml')
```

```python
actuals = {'tembotrione': 5.54,
      'fomesafen': 4.1,
      'atrazine': 1.0,
      'imazethapyr':7.8,
      'butafenacil':5.0,
      'nicosulfuron': 5.7,
      'dicamba': 6.1,
      'mesotrione': 5.1,
      'fluazifop': 4.45}

def dmsoDil(mw, w, c2 = 0.01, v2 = 1):
    target_weight = mgml(mw, c2) # for 1 ml
    frac_off = w / target_weight
    return frac_off * v2

for i, j in zip(mws, actuals):
    print(i, round(dmsoDil(mws[i], actuals[j]) * 1000, 3), 'l')
```
